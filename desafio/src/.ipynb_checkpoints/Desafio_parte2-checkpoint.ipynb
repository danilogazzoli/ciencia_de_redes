```python
import numpy as np
import random
import networkx as nx
from IPython.display import Image
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
G = nx.read_gml('../data/GraphMissingEdges.gml')

n = G.number_of_nodes()
m = G.number_of_edges()
print("Número de nós : %d" % n)
print("Número de arestas : %d" % m)
print("Número de componentes conectados : %d" % nx.number_connected_components(G))
```

```python
# Remove 20% das arestas
proportion_edges = 0.2
edge_subset = random.sample(G.edges(), int(proportion_edges * G.number_of_edges()))
```

```python
# Cria uma cópia do grafo e remove arestas
G_train = G.copy()
G_train.remove_edges_from(edge_subset)
```

```python
edge_subset_size = len(list(edge_subset))
print("Número de arestas removidas : %d" % edge_subset_size)
print("Número de arestas restantes : %d" % (m - edge_subset_size))
```

## Coefficiente de Jaccard

```python
# Fazendo previsões com o jaccard
pred_jaccard = list(nx.jaccard_coefficient(G_train))

score_jaccard = [s for (u,v,s) in pred_jaccard]
label_jaccard = [(u,v) in edge_subset for (u,v,s) in pred_jaccard]


# Computa o AUC score da curva ROC
fpr_jaccard, tpr_jaccard, _ = metrics.roc_curve(label_jaccard, score_jaccard)
auc_jaccard = roc_auc_score(label_jaccard, score_jaccard)


#Gráfico da curva ROC
plt.title('ROC')
plt.plot(fpr_jaccard, tpr_jaccard, 'b', label = 'AUC = %0.2f' % auc_jaccard)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('Taxa Verdadeiro Positivo')
plt.xlabel('Taxa de Falso Positivo')
plt.show()
```

## Preferential Attachment

```python
pred_prefAtch = list(nx.preferential_attachment(G_train))

score_prefAtch = [s for (u,v,s) in pred_prefAtch]

label_prefAtch = [(u,v) in edge_subset for (u,v,s) in pred_prefAtch]

# Computa o AUC score da curva ROC
fpr_prefAtch, tpr_prefAtch, _ = metrics.roc_curve(label_prefAtch, score_prefAtch)
auc_prefAtch = roc_auc_score(label_prefAtch, score_prefAtch)
print(auc_prefAtch)


#Gráfico da curva ROC
plt.title('ROC')
plt.plot(fpr_prefAtch, tpr_prefAtch, 'b', label = 'AUC = %0.2f' % auc_prefAtch)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('Taxa Verdadeiro Positivo')
plt.xlabel('Taxa de Falso Positivo')
plt.show()
```

```python
pred_prefAtch = list(nx.preferential_attachment(G, lista))
```

```python
lista_lontitude = []
lista_latitude = []
lista_valor = []
lista_p = []
```

```python
for u, v, p in pred_prefAtch:
    cate1 = G.nodes[u]['categories']
    lat1 = float(G.nodes[u]['latitude'])
    lon1 = float(G.nodes[u]['longitude'])
    
    
    cate2 = G.nodes[v]['categories']
    lat2 = float(G.nodes[v]['latitude'])
    lon2 = float(G.nodes[v]['longitude'])

    
    setCate1 = set(cate1.split(','))
    setCate2 = set(cate2.split(','))

    
    lista_valor.append(len(setCate1.intersection(setCate2)))
    lista_latitude.append(abs(lat1 - lat2))
    lista_lontitude.append(abs(lon1 - lon2))
    lista_p.append(p)
```

```python
latitude_limite = statistics.quantiles(lista_latitude)
longitude_limite = statistics.quantiles(lista_lontitude)
intersec_limite = statistics.quantiles(lista_valor)
intersec_p = statistics.quantiles(lista_p)
print(latitude_limite)
print(longitude_limite)
print(intersec_limite)
print(intersec_p)
```

```python
dfEdgesToEvaluate
```

```python
listInference = []
for quantidade_grupo,lat,lon,p in zip(lista_valor,lista_latitude,lista_lontitude,lista_p):
    # usando quartil superior para os q estão relativamente pertos e mesmo grupo então deve haver um link
    if quantidade_grupo >= intersec_limite[2] and lat < latitude_limite[2] and lon < longitude_limite[2]:
        listInference.append(1)# n pode abaixar
    else:
        # usando quartil inferior para os q estão muitos perto então deve haver um link
        if  lat < latitude_limite[0] and  lon < longitude_limite[0]: 
            listInference.append(1) 
        else:
            if p > intersec_p[2]:
                listInference.append(1)
            else:
                listInference.append(0)

dfEdgesToEvaluate['link'] = listInference
dfEdgesToEvaluate.to_csv("arestas.csv", columns=['linkID','link'],index=False)
```
